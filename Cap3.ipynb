# Capítulo 3 - Visualizando dados

```python
import matplotlib.pyplot as plt
from collections import Counter
```

## Linhas

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, gdp, color = "green", marker = 'o', linestyle = 'solid')
plt.title("PIB nominal")
plt.ylabel("Bilhões de dólares")
plt.show()
```

## Barras

```python
filmes = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
plt.bar(range(len(filmes)), num_oscars)
plt.title("Meus filmes favoritos")
plt.ylabel("Número de oscars")
plt.xticks(range(len(filmes)), filmes)
plt.show()
```

## Histograma

```python
notas = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
histograma = Counter(min(nota // 10 * 10, 90) for nota in notas)
```

```python
plt.bar([x + 5 for x in histograma.keys()],
        histograma.values(),
        10,
        edgecolor = (0, 0, 0))
plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decimais")
plt.ylabel("Número de estudantes")
plt.title("Distribuição das notas da 1 avaliação")
plt.show()
```

```python
mencoes = [500, 505]
anos = [2017, 2018]
```

```python
plt.bar(anos, mencoes, 0.8)
plt.xticks(anos)
plt.ylabel("Números de vezes que eu ouvi alguém falar 'data science'")
plt.ticklabel_format(useOffset = False)

plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Olha o quanto cresceu!")
plt.show()
```

```python
plt.bar(anos, mencoes, 0.8)
plt.xticks(anos)
plt.axis([2016.5, 2018.5, 0, 550])
plt.title("Não parece tão grande assim")
plt.show()
```

## Gráfico de linhas

```python
variancia = [2 ** x for x in range(9)]
vies_quadrado = [2 ** x for x in range(8, -1, -1)]
erro_total = [x + y for x, y in zip(variancia, vies_quadrado)]
xs = [i for i, _ in enumerate(variancia)]
```

```python
plt.plot(xs, variancia, 'g-', label = 'variancia')
plt.plot(xs, vies_quadrado, 'r-', label = 'vies^2')
plt.plot(xs, erro_total, 'b:', label = 'erro_total')
plt.legend(loc = 9)
plt.xlabel('Complexidade do modelo')
plt.xticks([])
plt.title("O balanço entre viés e variância")
plt.show()
```

## Disperção

```python
amizades = [70, 65, 72, 63, 71,64, 60, 64, 67]
minutos = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(amizades, minutos)

for label, amizade, minuto in zip(labels, amizades, minutos):
    plt.annotate(label,
                 xy = (amizade, minuto),
                 xytext = (5, -5),
                 textcoords = 'offset points')
    
plt.title("Minutos diários x Número de amigos")
plt.xlabel("Número de amigos")
plt.ylabel("Minutos por dia gastos no site")
plt.show()
```

```python
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Eixos não são comparáveis")
plt.xlabel("Notas da primeira prova")
plt.ylabel("Notas da segunda prova")
plt.show()
```
